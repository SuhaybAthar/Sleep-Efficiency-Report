```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Load the dataset from the CSV file
df = pd.read_csv('sleep_efficiency.csv')
```

```python
print('Hello')
```

```python
# Correlation matrix
correlation_matrix = df.corr()

# Figure 1: Heatmap for correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

```

```python
# Figure 2
# Scatter plot: Sleep Duration vs. Age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Sleep duration', data=df)
plt.title('Sleep Duration vs. Age')
plt.xlabel('Age')
plt.ylabel('Sleep Duration')
plt.show()
```

```python
# Figure 3
# Regression plot: Sleep Efficiency vs. Caffeine Consumption
plt.figure(figsize=(10, 6))
sns.regplot(x='Caffeine consumption_ml', y='Sleep efficiency', data=df)
plt.title('Sleep Efficiency vs. Caffeine Consumption')
plt.xlabel('Caffeine Consumption')
plt.ylabel('Sleep Efficiency')
plt.show()

# Statistical test: Correlation between Exercise Frequency and REM Sleep Percentage
exercise_rem_correlation, _ = pearsonr(df['Exercise frequency_weekly'], df['REM sleep percentage'])
print(f'Correlation between Exercise Frequency and REM Sleep Percentage: {exercise_rem_correlation:.2f}')
```

```python
# Figure 4: Smoking and Sleep Efficiency
plt.figure(figsize=(12, 6))
sns.boxplot(x='Smoking status', y='Sleep efficiency', data=df)
plt.title('Impact of Smoking on Sleep Efficiency')
plt.xlabel('Smoking Habits')
plt.ylabel('Sleep Efficiency')
plt.show()

# Figure 5: Alcohol Consumption and Sleep Efficiency
plt.figure(figsize=(12, 6))
sns.boxplot(x='Alcohol consumption_drinks', y='Sleep efficiency', data=df)
plt.title('Impact of Alcohol Consumption on Sleep Efficiency')
plt.xlabel('Alcohol Consumption (units)')
plt.ylabel('Sleep Efficiency')
plt.show()

# Figure 6: Caffeine Intake and Sleep Efficiency
plt.figure(figsize=(12, 6))
sns.boxplot(x='Caffeine consumption_ml', y='Sleep efficiency', data=df)
plt.title('Impact of Caffeine Intake on Sleep Efficiency')
plt.xlabel('Caffeine Intake (mg)')
plt.ylabel('Sleep Efficiency')
plt.show()

# Figure 7: Average Sleep Stage Percentages
avg_sleep_stages = df[['REM sleep percentage', 'Deep sleep percentage', 'Light sleep percentage']].mean()

plt.figure(figsize=(12, 8))
avg_sleep_stages.plot(kind='bar', color=['skyblue', 'royalblue', 'steelblue'])
plt.title('Average Sleep Stage Percentages')
plt.xlabel('Sleep Stages')
plt.ylabel('Average Percentage of Total Sleep')
plt.xticks(rotation=0)
plt.show()

```
